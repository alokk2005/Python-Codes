```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)             # Line plot
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.show()
```

```python
x = np.linspace(0, 10, 20)
y = np.sin(x)

plt.plot(x, y, color='red', linestyle='--', marker='o', label='Sine Wave')
plt.legend()
plt.show()
```

```python
y2 = np.cos(x)

plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
plt.show()
```

```python
plt.subplot(2, 1, 1)   # (rows, cols, index)
plt.plot(x, y, 'r')
plt.title("Sine")

plt.subplot(2, 1, 2)
plt.plot(x, y2, 'b')
plt.title("Cosine")
plt.tight_layout()
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 10]
plt.bar(x, y, color='blue')
plt.title("Bar Chart")
plt.show()
```

```python
data = np.random.randn(1000)   # Normal distribution
plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.title("Histogram")
plt.show()
```

```python
sizes = [20, 30, 25, 25]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
plt.title("Pie Chart")
plt.show()
```

```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title("Heatmap")
plt.show()
```

```python
import pandas as pd

data = {'Year': [2018, 2019, 2020, 2021],
        'Sales': [100, 120, 90, 150]}
df = pd.DataFrame(data)

df.plot(x='Year', y='Sales', kind='line', marker='o', title="Sales Over Years")
plt.show()
```

```python
study_hours = [1,2,3,4,5,6,7,8,9]
scores = [35, 50, 45, 60, 70, 65, 80, 85, 90]

colors = np.random.rand(9)
sizes = 100 * np.random.rand(9)

plt.scatter(study_hours, scores, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.title("Study Hours vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)   # x values
y = 2*x + 3                     # equation of straight line (y = 2x + 3)

plt.plot(x, y, label="y = 2x + 3")
plt.axhline(0, color="black")   # x-axis
plt.axvline(0, color="black")   # y-axis
plt.legend()
plt.title("Straight Line")
plt.show()
```

```python
plt.axhline(y=5, color="red", linestyle="--", label="y = 5")
plt.axvline(x=2, color="blue", linestyle=":", label="x = 2")
plt.legend()
plt.show()
```

```python
x = [1, 4]
y = [2, 8]
plt.plot(x, y, marker="o")
plt.title("Line between (1,2) and (4,8)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)
y = x**2   # y = x²

plt.plot(x, y, label="y = x²")
plt.legend()
plt.title("Quadratic Curve")
plt.show()
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Actual labels (ground truth)
y_true = [0, 0, 1, 1, 0, 1, 0, 1]

# Predicted labels (from model)
y_pred = [0, 1, 1, 1, 0, 0, 0, 1]

# Create confusion matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

# Visualize it
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
disp.plot(cmap="Blues")
plt.show()
```
